```python
import pandas as pd

# Load the dataset into a pandas DataFrame
df = pd.read_csv("C://Users//bsind//Downloads//Housing.csv")

# Display the first 5 rows
df.head()
```

```python
# Check for missing values in the dataset
missing_values = df.isnull().sum()

# Summarize the dataset (data types, number of rows/columns)
dataset_summary = df.info()

# Display missing values and dataset summary
print("Missing Values:\n", missing_values)
print("\nDataset Summary:\n", dataset_summary)
```

```python
# Generate descriptive statistics for numerical columns
basic_statistics = df.describe()

# Display basic statistics
print("Basic Statistics:\n", basic_statistics)
```

```python
# Select only the numeric columns in the DataFrame
numeric_columns = df.select_dtypes(include=['number']).columns

# Calculate the correlation matrix for numeric columns only
correlation = df[numeric_columns].corr()

# Create a heatmap for the correlation matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Scatter plot: Price vs. Area
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="area", y="price", hue="furnishingstatus", style="airconditioning", palette="Set2")
plt.title("Price vs. Area (Colored by Furnishing Status & AC Availability)", fontsize=14)
plt.xlabel("Area (sq. ft.)", fontsize=12)
plt.ylabel("Price ()", fontsize=12)
plt.legend(title="Furnishing Status & AC", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# Bar plot: Number of properties by stories
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x="stories", palette="muted")
plt.title("Number of Properties by Stories", fontsize=14)
plt.xlabel("Stories", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

```

According to the above bar plot we can find that In the data, two storied buildings are more in number.
